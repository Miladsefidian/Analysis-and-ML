```python
import numpy as np
import matplotlib.pyplot as plt

#sigmoid Function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

#Generate Values
z = np.linspace(-10,10,100)
sigmoid_values = sigmoid(z)

#plot
plt.plot(z,sigmoid_values)
plt.title('sigmoid function')
plt.xlabel('z')
plt.ylabel('sigmoid value')
plt.grid()
plt.show()
```
