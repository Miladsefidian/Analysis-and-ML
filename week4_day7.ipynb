```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
import numpy as np
from sklearn.linear_model import LinearRegression

#inspect data
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
df = pd.read_csv(url)

del df['sex']
del df['smoker']
del df['day']
del df['time']
# Calculate the correlation matrix
correlation_matrix = df.corr(numeric_only=True)

# Create the heatmap with a title
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

#Seperate Data by Gender
male_tips = df[df['sex'] == 'Male']['tip']

female_tips = df[df['sex'] == 'Female']['tip']

#perform t-test
t_stat, p_value = ttest_ind(male_tips, female_tips)

print("T- Statistic =", t_stat

print("P-value =", p_value)

#Interet result
alpha = 0.05
if p_value < alpha:
    print("Something is wrong with your data.")

#define variables
X = df['total_bill'].values.reshape(-1, 1)
Y = df['tip'].values

#Fit linear Regression
model = LinearRegression()

model.fit(X, Y)

#output coefficients
print('Slope:', model.coef_[0])
print('Intercept:', model.intercept_)
print('R-squared:', model.score(X, Y))

#Plot regression


sns.scatterplot(x=df['total_bill'], y=df['tip'], color='red')
plt.plot(df['total_bill'], model.predict(X), color='blue', label='data')
plt.title('Scatter plot of total bill vs tip r-squared')
plt.xlabel('Total bill')
plt.ylabel('Tip')
plt.legend()
plt.show()







```
